## Problem Statement

Investors and homeowners based in the city of Ames are looking for professionals to aid them in getting the best possible deals when buying/selling housing properties. As part of a real estate agency in Ames looking to boost annual sales, we want to develop the best regression model to predict housing prices and provide expert advice to investors and homeowners on which housing components and features would greatly influence housing prices.

## Executive Summary

This project is part of a [Kaggle challenge](https://www.kaggle.com/c/dsi-us-6-project-2-regression-challenge/overview) and explores the Ames housing data which contains the values for individual residential properties sold in Ames from 2006 to 2010. The dataset also includes information on many housing components and features corresponding to each sale.

Exploratory data analysis revealed strong correlation between the sale price and several housing components. Based on Pearson's correlation coefficient, we were able to segregate the more important features from the not so important ones. We also found there was strong correlation between some of the dependent variables which results in multicollinearity if used together in the regression model. To fix this, the variables were either combined or dropped altogether, leaving the one with higher correlation with the target variable. Many of the features are altered to be more interpretable and categorical data were one-hot encoded to make it readable for the regression model.

Feature engineering was done, where we did a log transformation to the target variable to get a more homoscedastic model as observed through residual plots before and after the transformation. By doing so, the performance of the model in making predictions was improved. Conversely, we saw from scatterplots and correlation coefficients that doing the log tranformation on all the other variables was not necessary a good thing.

Lasso regression was used as a quick feature selection tool and was left with 20 features for model development. Model selection was done by comparing the estimates of the R2 and RMSE scores via the 5-fold cross validation method. We ended up with a Lasso regression model as the production model, with a R2 score of 0.86 and RMSE of 22841 on test(hold-out) data. Since the hold-out method was also adopted, we could compare scores between train and test(hold-out) data and found slight overfitting of the data. But the small difference was acceptable by our means.

The top features which influenced the price predictions the most were found to be the total area of the house, overall quality, neighborhood region of house location, interior finish of the house's garage, and the total number of rooms. Surprisingly, the house age and type of dwelling were not found to be in these top features, likely due to the seasonality of the market and other correlated features.

* **Pool QC** - Represents ordinal data for pool quality, the NaN values should mean there is no pool.

However, we see that there are different numbers of null values amongst the Bsmt features although they are supposed to indicate that there is no basement. In other words, these Bsmt features should have equal number of null values. We investigate further to find out if there are values missing at random.

```python
# Check row data where there are discrepancies in the Bsmt features
# also include the Bsmt features which are non-ordinal
Bsmt_features = ['Bsmt Qual', 'Bsmt Cond', 'Bsmt Exposure', 'BsmtFin Type 1', 'BsmtFin SF 1', 'BsmtFin Type 2', 'BsmtFin SF 2', 'Bsmt Unf SF', 'Total Bsmt SF', 'Bsmt Full Bath', 'Bsmt Half Bath']
df_train.loc[(df_train['Bsmt Exposure'].isnull()) & ~((df_train['Bsmt Exposure'].isnull()) & (df_train['BsmtFin Type 1'].isnull()) & (df_train['BsmtFin Type 2'].isnull()) & (df_train['Bsmt Cond'].isnull()) & (df_train['Bsmt Qual'].isnull())),
             Bsmt_features]
```

We observe that the NaN values in the "Bsmt Exposure" column do not tally with the values in the other Bsmt related columns, i.e. the other Bsmt related columns indicate that there is a basement, while the NaN values in "Bsmt Exposure" supposedly mean that there is no basement. We deduce that these values are missing completely at random (MCAR), and choose to impute the mode for the "Bsmt Exposure" column.

```python
# Value to impute
df_train['Bsmt Exposure'].mode()
```

```python
# Imputation of mode
df_train.loc[[1456, 1547, 1997], 'Bsmt Exposure'] = 'No'
```

```python
# Check null values for Bsmt ordinal features
df_train[['Bsmt Qual', 'Bsmt Cond', 'Bsmt Exposure', 'BsmtFin Type 1', 'BsmtFin Type 2']].isnull().sum().sort_values(ascending=False)
```

We will do the same checks for the extra null value in "BsmtFin Type 2".

```python
# Checking discrepancy for 'BsmtFin Type 2'
df_train.loc[(df_train['BsmtFin Type 2'].isnull()) & ~((df_train['Bsmt Exposure'].isnull()) & (df_train['BsmtFin Type 1'].isnull()) & (df_train['BsmtFin Type 2'].isnull()) & (df_train['Bsmt Cond'].isnull()) & (df_train['Bsmt Qual'].isnull())),
             Bsmt_features]
```

```python
# Mapping ordinal variables
df_train['Lot Shape'] = df_train['Lot Shape'].map({'IR3': 1, 'IR2': 2, 'IR1': 3, 'Reg': 4})
df_train['Utilities'] = df_train['Utilities'].map({'ELO': 1, 'NoSeWa': 2, 'NoSewr': 3, 'AllPub': 4})
df_train['Land Slope'] = df_train['Land Slope'].map({'Sev': 1, 'Mod': 2, 'Gtl': 3})
df_train['Exter Qual'] = df_train['Exter Qual'].map({'Po': 1, 'Fa': 2, 'TA': 3, 'Gd':4, 'Ex':5})
df_train['Exter Cond'] = df_train['Exter Cond'].map({'Po': 1, 'Fa': 2, 'TA': 3, 'Gd':4, 'Ex':5})
df_train['Bsmt Qual'] = df_train['Bsmt Qual'].map({'Po': 1, 'Fa': 2, 'TA': 3, 'Gd': 4, 'Ex': 5})
df_train['Bsmt Cond'] = df_train['Bsmt Cond'].map({'Po': 1, 'Fa': 2, 'TA': 3, 'Gd': 4, 'Ex': 5})
df_train['Bsmt Exposure'] = df_train['Bsmt Exposure'].map({'Po': 1, 'Fa': 2, 'TA': 3, 'Gd': 4, 'Ex': 5})
df_train['BsmtFin Type 1'] = df_train['BsmtFin Type 1'].map({'Unf': 1, 'LwQ': 2, 'Rec': 3, 'BLQ': 4, 'ALQ': 5, 'GLQ': 6})
df_train['BsmtFin Type 2'] = df_train['BsmtFin Type 2'].map({'Unf': 1, 'LwQ': 2, 'Rec': 3, 'BLQ': 4, 'ALQ': 5, 'GLQ': 6})
df_train['Heating QC'] = df_train['Heating QC'].map({'Po': 1, 'Fa': 2, 'TA': 3, 'Gd': 4, 'Ex': 5})
df_train['Electrical'] = df_train['Electrical'].map({'Mix': 1, 'FuseP': 2, 'FuseF': 3, 'FuseA': 4, 'SBrkr': 5})
df_train['Kitchen Qual'] = df_train['Kitchen Qual'].map({'Po': 1, 'Fa': 2, 'TA': 3, 'Gd': 4, 'Ex': 5})
df_train['Functional'] = df_train['Functional'].map({'Sal': 1, 'Sev': 2, 'Maj2': 3, 'Maj1': 4, 'Mod': 5, 'Min2': 6, 'Min1': 7, 'Typ': 8})
df_train['Fireplace Qu'] = df_train['Fireplace Qu'].map({'Po': 1, 'Fa': 2, 'TA': 3, 'Gd': 4, 'Ex': 5})
df_train['Garage Finish'] = df_train['Garage Finish'].map({'Unf': 1, 'RFn': 2, 'Fin': 3})
df_train['Garage Qual'] = df_train['Garage Qual'].map({'Po': 1, 'Fa': 2, 'TA': 3, 'Gd': 4, 'Ex': 5})
df_train['Garage Cond'] = df_train['Garage Cond'].map({'Po': 1, 'Fa': 2, 'TA': 3, 'Gd': 4, 'Ex': 5})
df_train['Paved Drive'] = df_train['Paved Drive'].map({'N': 1, 'P': 2, 'Y': 3})
df_train['Pool QC'] = df_train['Pool QC'].map({'Fa': 1, 'TA': 2, 'Gd': 3, 'Ex': 4})
df_train['Fence'] = df_train['Fence'].map({'MnWw': 1, 'GdWo': 2, 'MnPrv': 3, 'GdPrv': 4})
```

#### Transforming Year variables  
We want to transform all the variables with the Year attribute to have more meaningful values. We will use the "Year Built" as a reference and transform the Year variables in relation to the reference year. For e.g., "Year Built" will be used to calculate "Age Till Sale" where it represents the total number of years since the house has been built till the time it was sold.

```python
year_features = ['Year Built', 'Year Remod/Add', 'Garage Yr Blt', 'Mo Sold', 'Yr Sold']
```

```python
# Study the year features first before transformation
df_train[year_features].describe()
```

```python
# Investigate anomaly
df_train[year_features].sort_values(by='Garage Yr Blt', ascending=False).head(1)
```

We notice an anomaly in the "Garage Yr Blt" column. Looking at the values of the other year features in the row data with the anomaly, it is safe to assume that it was a wrong data entry and the value should be 2007 instead. Also because the year of house remodeling is 2007, likely due to the garage being built.

```python
# Impute value
df_train.loc[[1699], ['Garage Yr Blt']] = 2007
```

```python
# Transforming the year features and renaming columns appropriately
df_train['Age Till Sale'] = df_train['Yr Sold'] + df_train['Mo Sold']/12 - df_train['Year Built']
df_train['Age Remod/Add Till Sale'] = df_train['Yr Sold'] + df_train['Mo Sold']/12 - df_train['Year Remod/Add']
df_train['Garage Age Till Sale'] = df_train['Yr Sold'] + df_train['Mo Sold']/12 - df_train['Garage Yr Blt']
year_features.append('Age Till Sale')
year_features.append('Age Remod/Add Till Sale')
year_features.append('Garage Age Till Sale')
df_train[year_features].head()
```

```python
# Describe the year features after transformation
df_train[year_features].describe()
```

```python
# Investigate anomaly
df_train[year_features].sort_values(by='Age Till Sale').head(1)
```

We notice an anomaly in the new feature columns, where the age is negative. Although it is possible that this house was sold even before it was built, for the sake of analysis in this project we will set all the age to zero, with the assumption that the house and its garage was built/remodeled at the same time it was sold.

```python
# Impute value
df_train.loc[[1885], ['Age Till Sale', 'Age Remod/Add Till Sale', 'Garage Age Till Sale']] = 0
```

```python
# Replace the NaN values with zero
df_train.loc[df_train['Garage Yr Blt'].isnull(), ['Garage Yr Blt', 'Garage Age Till Sale']] = 0
```

```python
# Checking dataframe null values
leftover_nulls = df_train.isnull().sum()[df_train.isnull().sum()>0]
leftover_nulls
```

```python
# Function to plot multiple violinplots
def subplot_violinplot(dataframe, x_list, y_target):
    nrows = int(np.ceil(len(x_list)/3))   # Makes sure you have enough rows
    fig, ax = plt.subplots(nrows=nrows, ncols=3, figsize=(14, nrows*4))   
#     plt.subplots_adjust(hspace = 0.4, wspace=0.3)
    ax = ax.ravel()   # Ravel turns a matrix into a vector, which is easier to iterate
    for i, column in enumerate(x_list):   # Gives us an index value to get into all our lists
        sns.violinplot(x=column, y=y_target, data=dataframe, ax=ax[i], palette='Accent')
    
        # display scale and label only on the left edge
        if i%3 != 0:
            ax[i].set_ylabel('')
            ax[i].set_yticklabels('')
    # This code below is to switch off the axes if there is an odd number of subplots
    for axes in ax.flat[len(x_list):]:
        axes.axis('off')
    fig.tight_layout()
    pass
```

```python
# Look at ordinal features
ord_var
```

```python
# Zoom into the top 5 features with highest correlation with SalePrice
# We choose top 5 here because these have correlation above 0.5
n = 6
corr_var = ord_var.copy()
corr_var.append('SalePrice')
top_num_col = abs(df_train[corr_var].corr()).nlargest(n, 'SalePrice').index   # extract the columns for the top 10
corrmat = df_train[top_num_col].corr()
mask_matrix = np.triu(corrmat)

plt.figure(figsize=(5,5))
plt.title('Heatmap of Top 5 Correlation for Ordinal Variables', fontsize=10)
sns.heatmap(corrmat, cmap='OrRd', mask=mask_matrix, square=True, annot=True);
```

We plotted a heatmap of the features among the top 5 correlation coefficients with "SalePrice". These are the features that we will be keeping to develop the first prediction model. However, out of these features, there are features with high correlation (r > 0.6) and may result in multicollinearity. We will plot a violinplot to make a more informed decision on which features to keep/drop.

```python
# Plot violinplots for top 5 ordinal features
plot_list = [col for col in top_num_col if col in ord_var]
subplot_violinplot(df_train, plot_list, 'SalePrice')
```

Based on both the heatmap and the violinplot, we can choose which features to keep/drop. From the violinplots, we are also able to observe if the ordinal variable exhibit ordinal characteristics. ## Encode or Map Categorical Data

For categorical/nominal data, we will be primarily be using the one-hot encoding method. For each feature, we will drop one of the dummy columns and take that as the baseline reference for the other dummy columns (i.e. if all the dummy columns in a feature is 0, it is the baseline reference). Generally, we will choose the dummy column with the lowest value count to drop so that the features will be more representative.

```python
# Dataframe before encoding
df_train.shape
```

```python
# One-hot encoding using get_dummies
df_train = pd.get_dummies(df_train)
df_train.head()
```

* Combined Exterior_WoodMetal

```python
# Dropping the chosen dummy columns
df_train.drop(columns=['MS SubClass_TwoHalfStorey', 'Neighborhood_Lincoln Way/State Ave', 'Foundation_Wood', 'Combined Exterior_WoodMetal'],
             inplace=True)
```
